exang - An indication using 1 or 0 if exercise is the cause of angina, a type of chest pain. Which can be an indication of cardiovascular disease. 

oldpeak - A form of ST depression that is influenced or caused by exercising in relative to how much they rest. An ST depression in an electriccardiograph may indicate severe coronoary lesions and unstable coronary artery disease. An ST depression can be found when the ST segment is abnormally low below the normal baseline in an ECG test. The oldpeak is measured by the distance form the baseline to the depression segment lowest point.

slope = the slope of the peak exercise ST segment

ST = There is a brief period between the end of the QRS complex and the beginning of the T wave where there is no conduction and the line is flat. This is known as the ST segment and it is a key indicator for both myocardial ischaemia and necrosis if it goes up or down. The ST segment is the flat, isoelectric section of the ECG between the end of the S wave (the J point) and the beginning of the T wave. The ST Segment represents the interval between ventricular depolarization and repolarization. The most important cause of ST segment abnormality (elevation or depression) is myocardial ischaemia or infarction. ST-segment elevation usually indicates a total blockage of the involved coronary artery and that the heart muscle is currently dying. Non-STEMI heart attacks usually involve an artery with partial blockage, which usually does not cause as much heart muscle damage.

ca =` number `of major vessels (0-3) colored by flouroscopy

fluo
roscopy = Fluoroscopy is an imaging technique that uses X-rays to obtain real-time moving images of the interior of an object.



Cp \- 4 values of chest pain; 1 = asymptomatic, 2 = non angina pain \(pain in chest/heart that is non cardiac\), 3 = typical type angina, 4= typical type. 

thest bps \- sleeping blood force \(mm hg or millimeters of mercury\)



Chol \- Serum cholesterol in mg/dL \(In healthy people generally less than 200 mg/dL\)

fbs \- fasting blood sugar &gt; 120 mg/dl \(1 indicates greater than 120 mg/dL, 0 indicates below\) \(Normal is 99 mg/dL or lower, 100\-125 indicates prediabetes, 126\+ indicates diabetes\)



thal \- thalassemia; blood disorder that makes your body have less hemoglobin than usual \(can make you tired\); 0 = normal blood flow, 1 = fixed defect \(no blood flow in certain parts of the heart\), 2 = reversible defect \(blood flow is being observed but it normal\)
